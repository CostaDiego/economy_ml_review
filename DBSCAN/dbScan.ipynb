```python
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

iris = load_iris()
```

```python
## Plot com as variáveis de largura e comprimento das pétalas, cada cor é um tipo de flor.
plt.scatter(iris["data"][:, 0], iris["data"][:, 1], c = iris["target"])
```

```python
## Ajuste do modelo
X = iris["data"][:, 0:2]
dbScan = DBSCAN(eps = 0.0375, min_samples = 3)
dbScan.fit(X)
```

```python
## Plot com os clusters formados pelo DBSCAN
plt.scatter(X[:, 0], X[:, 1], c= dbScan.labels_, s=50, cmap='viridis')
```
